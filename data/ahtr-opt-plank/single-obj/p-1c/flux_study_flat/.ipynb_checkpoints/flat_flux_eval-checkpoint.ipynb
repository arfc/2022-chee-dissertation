```python
import openmc
import openmc.mgxs as mgxs
import sys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sp = openmc.StatePoint('./statepoint.80.h5')
```

```python
tally_flux_all = sp.get_tally(name='flux_all')
flux_all = tally_flux_all.get_slice(scores=['flux'])
nu_fission_all = tally_flux_all.get_slice(scores=['nu-fission'])
fission_all = tally_flux_all.get_slice(scores=['fission'])
nu = sum(nu_fission_all.mean) / sum(fission_all.mean)
Q = 200 * 1.6022e-13
V = 27.1 * 3.25 * 1 / (100 * 20)
k = sp.k_combined.n
P = 1456
N = P * nu / (Q * k)
flux_all_vals = flux_all.mean * 1 / V * N
flux_all_vals.shape = (20, 100)
```

```python
fig, ax = plt.subplots(figsize=(7,5))
ax.plot(flux_all_vals[10])
ax.set_ylim(2.5e14, 3.5e14)
ax.grid()
```

```python
def get_values(sp_file):
    tally_flux = sp_file.get_tally(name='flux')
    flux = tally_flux.get_slice(scores=['flux'])
    nu_fission = tally_flux.get_slice(scores=['nu-fission'])
    fission = tally_flux.get_slice(scores=['fission'])
    flux_conv_dict = {}
    eg_names = ['eg4', 'eg3', 'eg2', 'eg1']
    egs = [(1e-6, 1.8554), (1.8554, 2.9023e1), (2.9023e1, 9.1188e3), (9.1188e3, 2.0e7)]
    flux_eg, nu, fiss_eg = {}, {}, {}
    for x in range(4):
        print(x)
        flux_eg[x] = flux.get_slice(
            filters=[
                openmc.EnergyFilter], filter_bins=[
                (egs[x],)])
        nu_fiss_eg = nu_fission.get_slice(
            filters=[
                openmc.EnergyFilter], filter_bins=[
                (egs[x],)])
        fiss_eg[x] = fission.get_slice(
            filters=[
                openmc.EnergyFilter], filter_bins=[
                (egs[x],)])
        nu[x] = sum(nu_fiss_eg.mean) / sum(fiss_eg[x].mean)
    return flux_eg, nu, fiss_eg
```

```python
flux_eg, nu, fiss_eg = get_values(sp)
```

```python
def plot_fission(group):
    x = np.arange(0, 27.1, 27.1/100)
    fission_eg = np.array(fiss_eg[group].mean)
    fission_eg.shape = (20, 100)
    fig, ax = plt.subplots(figsize=(7,5))
    ax.plot(x, fission_eg[10])
    #ax.set_ylim(bot, top)
    ax.set_xlabel('x [cm]', fontsize=16)
    ax.set_ylabel(r'Fission RR', fontsize=16)
    ax.set_title('Group '+ str(group+1))
    ax.grid()
    ax.set_xlim(0, 27.1)
```

```python
plot_fission(0)
plot_fission(1)
plot_fission(2)
plot_fission(3)
```

```python
def get_flux_dict_plank(k, P, z_thickness):
    Q = 200 * 1.6022e-13
    V = 27.1 * 3.25 * z_thickness / (100 * 20)
    flux_conv_dict = {}
    eg_names = ['eg4', 'eg3', 'eg2', 'eg1']
    for x in range(4):
        N = P * nu[x] / (Q * k)
        flux_conv_dict[eg_names[x]] = flux_eg[x].mean * 1 / V * N
        flux_conv_dict[eg_names[x]].shape = (20, 100)
        flux_conv_dict[eg_names[x]][np.isnan(flux_conv_dict[eg_names[x]])] = 0
    return flux_conv_dict
```

```python
power_moltres = 1456 # 1cm thick
flux_vals_openmc = get_flux_dict_plank(sp.k_combined.n, power_moltres, z_thickness=1)
```

```python
def plot_flux(flux_vals, group, bot, top):
    x = np.arange(0, 27.1, 27.1/100)
    flux = np.array(flux_vals['eg'+str(group)][10])
    #print(flux)
    fig, ax = plt.subplots(figsize=(7,5))
    ax.plot(x, flux)
    ax.set_ylim(bot, top)
    ax.set_xlabel('x [cm]', fontsize=16)
    ax.set_ylabel(r'Flux [$n/cm^2s$]', fontsize=16)
    ax.set_title("group "+str(group))
    ax.grid()
    ax.set_xlim(0, 27.1)
```

```python
plt.figure()
plt.imshow(flux_vals_openmc["eg4"]/np.mean(flux_vals_openmc["eg1"]),
           interpolation='none',origin='lower',cmap='viridis')
```

```python
plot_flux(flux_vals_openmc, 1, 0.8e14, 1.2e14)
plot_flux(flux_vals_openmc, 2, 0.6e14, 1.0e14)
plot_flux(flux_vals_openmc, 3, 0.6e13, 0.6e14)
plot_flux(flux_vals_openmc, 4, 0.8e14, 1.6e14) 
```

```python
def total_perc(flat):
    total_flat = 0
    perc_flat = {}
    for i in range(1,5):
        total_flat += sum(sum(flat["eg"+str(i)]))
    for i in range(1,5):
        perc_flat["eg"+str(i)] = sum(sum(flat["eg"+str(i)])) / total_flat * 100
    return perc_flat
```

```python
def get_fiss_dict(fiss_eg):
    eg_names = ['eg4', 'eg3', 'eg2', 'eg1']
    fiss_dict = {}
    for x in range(4):
        fiss_dict[eg_names[x]] = fiss_eg[x].mean
        fiss_dict[eg_names[x]].shape = (100, 20)
    return fiss_dict
```
