```python
def diamond_points():
    d_points = """"""
    point = 22
    angles = [0, -2*pi/3, 2*pi/3]
    # left side
    for j in range(3):
        d_points += "\n// Diamond Points Left Area " + str(j+1) + "\n"
        x, y = x_22, y_22 
        for i in range(14):
            if (i %2) == 0:
                if i > 0:
                    x += pc.P_A1_adj
                    y -= pc.P_A1_height
            else:
                if i in [1, 13]:
                    x += pc.P_small_gap / tan(pi/3) 
                    y -= pc.P_small_gap
                else:
                    x += pc.P_big_gap / tan(pi/3) 
                    y -= pc.P_big_gap
            d_points += "Point(%d) = {%f, %f, 0, 1.0}; \n" % (point, pc.rx(x, y, angles[j]), pc.ry(x, y, angles[j]))
            point += 1
    # right side 
    point = 64
    x_64 = -pc.D_to_center_width
    y_64 = -pc.D_to_center
    for j in range(3):
        d_points += "\n// Diamond Points Right Area " + str(j+1) + "\n"
        x, y = x_64, y_64 
        for i in range(14):
            if (i %2) == 0:
                if i > 0:
                    x += pc.P_A1_adj
                    y -= pc.P_A1_height
            else:
                if i in [1, 13]:
                    x += pc.P_small_gap / tan(pi/3) 
                    y -= pc.P_small_gap
                else:
                    x += pc.P_big_gap / tan(pi/3) 
                    y -= pc.P_big_gap
            d_points += "Point(%d) = {%f, %f, 0, 1.0}; \n" % (point, pc.rx(x, y, angles[j]), pc.ry(x, y, angles[j]))
            point += 1
    return d_points
```

```python
def stripe_points():
    stripe_points = """""" 
    angles = [0, -2*pi/3, 2*pi/3]
    # left side 
    point = 178 
    for j in range(3):
        stripe_points += "\n// Stripe Points Left Area " + str(j+1) + "\n"
        x_plank, y_plank = x_106, y_106 
        for i in range(12):
            x = x_plank + 0.5 * pc.P_len - 0.5 * pc.F_len
            if (i %2) == 0:
                y = y_plank - pc.F_protect_gap
                x_plank += pc.P_A1_adj
                y_plank -= pc.P_A1_height
            else:
                y = y_plank + pc.F_protect_gap
                x_plank += pc.P_big_gap / tan(pi/3) 
                y_plank -= pc.P_big_gap
            for k in range(2):
                if k > 0:
                    if (i %2) == 0:
                        y -= pc.F_width
                    else:
                        y += pc.F_width
                stripe_points += "Point(%d) = {%f, %f, 0, 1.0}; \n" % (point, pc.rx(x, y, angles[j]), pc.ry(x, y, angles[j]))
                point += 1
    # right side 
    point = 250 
    for j in range(3):
        stripe_points += "\n// Stripe Points Right Area " + str(j+1) + "\n"
        x_plank, y_plank = x_142, y_142 
        for i in range(12):
            x = x_plank - 0.5 * pc.P_len + 0.5 * pc.F_len
            if (i %2) == 0:
                y = y_plank - pc.F_protect_gap
                x_plank += pc.P_A1_adj
                y_plank -= pc.P_A1_height
            else:
                y = y_plank + pc.F_protect_gap
                x_plank += pc.P_big_gap / tan(pi/3) 
                y_plank -= pc.P_big_gap
            for k in range(2):
                if k > 0:
                    if (i %2) == 0:
                        y -= pc.F_width
                    else:
                        y += pc.F_width
                stripe_points += "Point(%d) = {%f, %f, 0, 1.0}; \n" % (point, pc.rx(x, y, angles[j]), pc.ry(x, y, angles[j]))
                point += 1
    
    return stripe_points
```

```python
def spacers_points():
    spacer_points = """"""
    angles = [0, -2*pi/3, 2*pi/3]
    size = [pc.S_small_r, pc.S_large_r, pc.S_large_r, pc.S_large_r, pc.S_large_r, pc.S_large_r, pc.S_small_r]
    # left side
    point = 322
    for j in range(3):
        x_plank, y_plank = x_106, y_106
        spacer_points += "\n// Spacer Points Area " + str(j+1) + "\n"
        for i in range(7):
            for k in range(2):
                if k > 0:
                    x_mid = x_plank + 0.5 * pc.P_len + 0.5 * pc.S_S_gap 
                else:
                    x_mid = x_plank + 0.5 * pc.P_len - 0.5 * pc.S_S_gap 
                y_mid = y_plank
                spacer_points += "Point(%d) = {%f, %f, 0, 1.0}; \n" % (point, pc.rx(x_mid, y_mid, angles[j]), pc.ry(x_mid, y_mid, angles[j]))
                point += 1
                x, y = x_mid - size[i], y_mid 
                spacer_points += "Point(%d) = {%f, %f, 0, 1.0}; \n" % (point, pc.rx(x, y, angles[j]), pc.ry(x, y, angles[j]))
                point += 1
                if i == 6:
                    x, y = x_mid, y_mid - size[i]
                else:
                    x, y = x_mid, y_mid + size[i]
                spacer_points += "Point(%d) = {%f, %f, 0, 1.0}; \n" % (point, pc.rx(x, y, angles[j]), pc.ry(x, y, angles[j]))
                point += 1
                x, y = x_mid + size[i], y_mid 
                spacer_points += "Point(%d) = {%f, %f, 0, 1.0}; \n" % (point, pc.rx(x, y, angles[j]), pc.ry(x, y, angles[j]))
                point += 1
            if i == 5:
                x_plank += (pc.P_A1_height)/ tan(pi/3)
                y_plank -= (pc.P_A1_height)
            else:
                x_plank += (pc.P_A1_height + pc.P_big_gap)/ tan(pi/3)
                y_plank -= (pc.P_A1_height + pc.P_big_gap)
    return spacer_points
```

```python
def spacer_lines_surfaces():
    spacer_lines = """"""
    spacer_surfaces = """"""
    line = 166
    point = 322
    loop = 40
    surfaces = ""
    for j in range(3):
        spacer_lines += "\n// Spacer Lines Area " + str(j+1) + "\n"
        spacer_surfaces += "\n// Spacer Surfaces Area " + str(j+1) + "\n"
        for i in range(14):
            spacer_lines += "Circle(%d) = {%d, %d, %d}; \n" % (line, point+2, point, point+1)
            line += 1
            spacer_lines += "Line(%d) = {%d, %d}; \n" % (line, point+1, point+3)
            line += 1
            spacer_lines += "Circle(%d) = {%d, %d, %d}; \n" % (line, point+3, point, point+2)
            spacer_surfaces += "Curve Loop(%d) = {%d, %d, %d}; \n" % (loop, line-2, line-1, line)
            spacer_surfaces += "Plane Surface(%d) = {%d}; \n" % (loop-1, loop)
            if j == 2 and i == 13:
                surfaces += str(loop-1)
            else:
                surfaces += str(loop-1) + ", "
            line += 1
            loop += 1
            point += 4
    spacer_surfaces += "Physical Surface('spacers') = {%s}; \n" % (surfaces)
    return spacer_lines + spacer_surfaces 
```

```python
def extra_points():
    extra_points = """"""
    angles = [0, -2*pi/3, 2*pi/3]
    total_dx = 22
    dx = pc.D_A1_width/total_dx
    point = 490
    surface = 81
    x_start, y_start = x_22, y_22 - pc.P_small_gap/2
    for k in range(3):
        extra_points += "\n// Extra Points Area " + str(k+1) + "\n"
        x, y = x_start, y_start
        for j in range(7):
            dx_x = x
            for i in range(1, total_dx):
                dx_x += dx
                if i not in [4, 18]:
                    extra_points += "Point(%d) = {%f, %f, 0, 1.0}; \n" % (point, pc.rx(dx_x, y, angles[k]), pc.ry(dx_x, y, angles[k]))
                    if i in [5, 19]:
                        surface += 1
                    extra_points += "Point{%d} In Surface{%d}; \n" % (point, surface)
                    point += 1
            surface += 1
            if j in [0, 5]:
                x += (pc.P_small_gap/2 + pc.P_A1_height+ pc.P_big_gap/2) / tan(pi/3)
                y -= (pc.P_small_gap/2 + pc.P_A1_height+ pc.P_big_gap/2)
            else:
                x += (pc.P_A1_height+ pc.P_big_gap) / tan(pi/3)
                y -= (pc.P_A1_height+ pc.P_big_gap) 
    return extra_points 
```
